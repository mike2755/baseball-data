```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

mlb = pd.read_excel('/home/mike/Desktop/Pitt/Teams.xlsx', sheetname ='Sheet2')
y = mlb['attendance']
x = mlb['W']
x = sm.add_constant(x)
model = sm.OLS(y,x).fit()
plt.figure()
plt.scatter(mlb['W'],mlb['attendance'])
plt.xlabel("Wins (All teams from 2000-2018)")
plt.ylabel("Attendance")
plt.title("Wins vs. Attendance")


mlb['Playoff_Div'] = np.where((mlb.DivWin == 'Y'),1,0)
mlb['Playoff_WC'] = np.where(mlb.WCWin =='Y', 1,0)
mlb['PO'] = mlb.Playoff_Div + mlb.Playoff_WC

mlbPO = mlb.query('PO == 1')
plt.figure()
plt.scatter(mlbPO['W'],mlbPO['attendance'])
plt.xlabel("Wins (Playoff teams from 2000-2018)")
plt.ylabel("Attendance")
plt.title("Wins vs. Attendance for playoff teams")
Y= mlbPO['attendance']
X= mlbPO[['W']]
X = sm.add_constant(X)
modelPO = sm.OLS(Y,X).fit()
plt.show()
```
